```python
incident_15_25['DateOfCall'] = pd.to_datetime(incident_15_25['DateOfCall'])
incident_15_25['TimeOfCall'] = pd.to_datetime(incident_15_25['TimeOfCall'], format='%H:%M:%S')
```

```python
incident_15_25['TimeOfCall'] = incident_15_25['TimeOfCall'].dt.time
```

Gestion des données textuelles

```python
import numpy as np
text_cols = [ 'IncidentGroup', 'StopCodeDescription', 'SpecialServiceType', 'PropertyCategory', 'IncGeo_BoroughName', 'IncGeo_WardNameNew' ]

for col in text_cols:
    if incident_15_25[col].dtype == 'object':
        lower_unique = len(incident_15_25[col].dropna().str.lower().unique())
        raw_unique = incident_15_25[col].nunique(dropna=True)
        if lower_unique < raw_unique:
            print(f"{col} → doublons liés à la casse détectés")

# Nettoyage et uniformisation de la colonne IncGeo_WardNameNew
incident_15_25['IncGeo_WardNameNew'] = (
    incident_15_25['IncGeo_WardNameNew']
    .astype(str)
    .str.title())

# Remplace le texte "Nan" par de vraies valeurs manquantes
incident_15_25['IncGeo_WardNameNew'].replace('Nan', np.nan, inplace=True)

# Supprime les doublons causés par des différences de casse
incident_15_25['IncGeo_WardNameNew'] = incident_15_25['IncGeo_WardNameNew'].replace(
    incident_15_25['IncGeo_WardNameNew'].unique(),
    pd.Series(incident_15_25['IncGeo_WardNameNew'].unique()).str.title().values)
```

Création de variables dérivées

```python
# Dérivation de variables temporelles
incident_15_25['MonthOfCall'] = incident_15_25['DateOfCall'].dt.month
incident_15_25['DayOfWeek'] = incident_15_25['DateOfCall'].dt.day_name()
incident_15_25['IsWeekend'] = incident_15_25['DayOfWeek'].isin(['Saturday', 'Sunday'])
incident_15_25['DateAndTimeOfCall'] = pd.to_datetime(incident_15_25['DateOfCall'].astype(str) + ' ' + incident_15_25['TimeOfCall'].astype(str), errors='coerce')

# Catégorisation du temps d'intervention
def categorize_response_time(x):
    if pd.isna(x):
        return np.nan
    elif x < 300:
        return '<5 min'
    elif x <= 600:
        return '5–10 min'
    else:
        return '>10 min'

incident_15_25['ResponseTimeCategory'] = incident_15_25['PumpMinutesRounded'].apply(categorize_response_time)
```

ANOVA

```python
import statsmodels.api

# H0 : pas d'effet significatif de la variable catégorielle sur la variable continue
# H1 : effet significatif de la variable catégorielle sur la variable continue

result = statsmodels.formula.api.ols('FirstPumpArriving_AttendanceTime ~ MonthOfCall', data = df_final).fit()
table = statsmodels.api.stats.anova_lm(result)

p_value = table['PR(>F)'][0]

print("Effet du mois sur le temps de réponse ?")
print("La p-value vaut :", p_value)
if p_value < 0.05:
    print("Effet sur la variable")
else:
    print("Pas d'effet sur la variable")
```
